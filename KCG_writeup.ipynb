### A simple model

For this strategy, we look at the price ratio between both instrument.
We postulate that, given the high correlation, the ratio should remain more or less constant over time and that variations are normally distributed. This means when the divergence hits a certain level, it is statistically more likely to revert back towards the mean.
Our strategy is to place a dollar-neutral trade at that point and revert it when the price ratio converges again.

````python
# try to establish what a 'normal' price ratio is
prc_ratio = df.prcA/df.prcB
prc_ratio_mean = prc_ratio.mean()
prc_ratio_std = prc_ratio.std()
# add this as part of the dataframe
df['ratio'] = prc_ratio
# setting up boundaries to enter a trade or revert it - they don't necessarily need to be identical
# note those were set arbitrarily - see later for a way to do this programatically
boundary_enter = prc_ratio_mean + 0.5*prc_ratio_std
boundary_revert = prc_ratio_mean + 0.1*prc_ratio_std

# inventory - inst: (<quantity>, <price bought at>)
from collections import namedtuple
Posn = namedtuple('Posn', ['qty', 'prc'])
inv = {'A': Posn(qty=0, prc=0), 'B': Posn(qty=0, prc=0)}

def is_inventory_empty(inv):
    A, B = inv['A'], inv['B']
    return A.qty == 0 and B.qty == 0

def inventory_value(inv):
    A, B = inv['A'], inv['B']
    return A.qty*A.prc + B.qty*B.prc
    
def calc_pnl(trades):
    pnl = 0
    for trade in trades:
        pnl += trade['A'].qty * trade['A'].prc
        pnl += trade['B'].qty * trade['B'].prc
        #print('running pnl: {0}'.format(pnl))
    return pnl

def apply_strategy(df, inv, boundary_revert, boundary_enter, verbose=False):
    # trades list
    trades = []
    for (ts, prcA, prcB, ratio) in df.itertuples():
        if boundary_enter < ratio and is_inventory_empty(inv):
            # short A & long B in a dollar-neutral way
            qtyA, qtyB = 1, ratio
            inv['A'] = Posn(qty=qtyA,prc=prcA)
            inv['B'] = Posn(qty=qtyB,prc=-prcB)
            trades.append( inv.copy() )
            if verbose:
                print('entered a trade: {0}'.format(inv))
                print('cost of trade (should be zero, or very close!): {0}'.format(inventory_value(inv)))
        elif ratio < boundary_revert and not is_inventory_empty(inv):
            A = inv['A']
            B = inv['B']
            # revert the trade by buying qtyA of A and selling qtyB of B
            Adash = Posn(qty=qtyA, prc=-prcA)
            Bdash = Posn(qty=qtyB, prc=prcB)

            trades.append( {'A':Adash, 'B':Bdash} )
            # reset the inventory since we're now flat
            # XXX in real life this might not be entirely feasible due to the inability to buy infinitely small amounts of an asset 
            inv = {'A': Posn(qty=0, prc=0), 'B': Posn(qty=0, prc=0)}

            if verbose:
                print('A was shorted at {0} and covered at {1}'.format(A.prc, Adash.prc))
                print('B was bought at {1} and sold at {2}, with qty {0}'.format(B.qty, B.prc, Bdash.prc))

    return trades

trades = apply_strategy(df, inv, boundary_revert, boundary_enter)
print('pnl: {0}'.format(calc_pnl(trades)))
````

### Improvements

Now that there is some basic understanding of what needs to be done, we can look at optimising the boundaries used to enter and revert a trade.

````python
slice_start = 0
slice_stop = len(df.index)
slice_mid = int(slice_stop/2) # as we're using this as an index, it can't be a float

df_training = df.iloc[slice_start:slice_mid].copy() # we create a copy because slicing returns a view!

# recreate the price ratio based on the training set
prc_ratio_tr = df_training.prcA/df_training.prcB
prc_ratio_mean_tr = prc_ratio_tr.mean()
prc_ratio_std_tr = prc_ratio_tr.std()

df_training['ratio'] = prc_ratio_tr

# let's see how the ratio is distributed
prc_ratio_tr.hist(bins=40)
````

Unlike what was assumed, this is not normally distributed and highlights how instrument B is sometimes more expensive than instrument A.

We want to find the boundary that divides the observations greater than the mean into two equal halves - that is, the median. We pick this as the optimal boundary given we want to remain as flat as often as possible so every trade can be unwound (we don't want to leave trades on our books for any longer than we have to). It won't maximise profits but should minimise the risk of us holding a costly position.

In practice however there'll likely be clustering.

````python
opt_boundary = prc_ratio_tr.median()
print('mean price ratio: {0}'.format(prc_ratio_tr.mean()))
print('optimal boundary: {0}'.format(opt_boundary))
inv = {'A': Posn(qty=0, prc=0), 'B': Posn(qty=0, prc=0)}

trades = apply_strategy(df_training, inv, boundary_revert=opt_boundary, boundary_enter=opt_boundary)
print('training set pnl: {0}'.format( calc_pnl(trades)) )

# we now do the same thing on our validation set
df_validation = df.iloc[slice_mid:slice_stop].copy() # we create a copy because slicing returns a view!

# recreate the price ratio based on the validation set
prc_ratio_v = df_validation.prcA/df_validation.prcB
df_validation['ratio'] = prc_ratio_v

inv = {'A': Posn(qty=0, prc=0), 'B': Posn(qty=0, prc=0)}
trades = apply_strategy(df_validation, inv, boundary_revert=opt_boundary, boundary_enter=opt_boundary)
print('validation set pnl: {0}'.format( calc_pnl(trades)) )
````

### Miscellaneous

#### Not Normal?

In a normal distribution, the mean and the median are identical. We originally assumed the price ratio was going to be normally distributed - however as our optimal boundary and histogram shows, it isn't the case.

#### Assumptions

We assumed there is always sufficient liquidity to place our trades to be executed at the requested level. We also assume frictionless trading (no fees, no taxes, no margin requirements etc...). We also assume holding a position at the end of the day is allowed.

#### Data Quality

A few liberties were taken with the data (such as forward filling or ignoring badly formatted data). There must be ways of building a more reliable dataset.

#### Dynamic boundaries

Looking at the moving average (or some sort of moving correlation), we might be able to set the boundary dynamically. That'd be something interesting to look at.

#### Correlation

In this sample, correlation is relatively constant. However we would need to monitor this, perhaps via some running correlation coefficient, to make sure the relationship still holds. Should it break, we would look to close our position as quickly as possible and re-assess.

#### Price averaging

Currently once a trade has been placed, we don't add to the position even if the spread moves in our favour. This means our entry point isn't always optimal. By extending the position we would be able to average out the cost of the 'good' trade.

#### Leverage

Right now we limit the strategy to taking on a single unit of instrument A. If we knew what the volumes were, we should be in a position to leverage accordingly.

#### Monitoring metrics

A useful metric might be to see how long we hold a position for before being able to unwind it. If being held for greater than a particular threshold, we could look to close it out as it might indicate our assumptions no longer hold.
